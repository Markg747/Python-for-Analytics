```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,6,8,10]
#create a simple line plot
plt.plot(x,y)
#adding labels
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Simple line chart')
#displaying the plot
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
#create a simple line plot
plt.plot(x,y,color='red',linestyle='--',marker='o',label='Line1')
plt.grid(True)
plt.legend()
```

```python
#invloves plotting multiple sets of data on the same axes

x=[1,2,3,4,5,6]
y1=[2,4,6,8,10,12]
y2=[3,5,7,9,11,13]
y3=[6,9,12,15,18,21]
#create multiline plot
plt.plot(x,y1,label='line1',marker='o')
plt.plot(x,y2,label='line2',marker='o')
plt.plot(x,y3,label='line3',marker='o')
#add label and title
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Multi Line Plot')
#add legend
plt.legend()
#display the plot
plt.show()
```

```python

import numpy as np
x=np.linspace(0,5,100)
y1=x**2
y2=2*x
#create a filled plot
plt.fill_between(x,y1,y2,color='pink',alpha=0.4,label= 'Fill Between')
#plot the lines
plt.plot(x,y1,label='line1')
plt.plot(x,y2,label='line2')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('fill  Plot')
plt.legend()
plt.show()
```

```python
#bar()
categories=['A','B','C','D']
values=[100,23,45,12]
#create a bar chart
plt.bar(categories,values,color='black')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart')
plt.legend()
plt.show()
```

```python

labels=['Java','C','C#','Python','Machine Learning']
sizes=[25,25,15,30,5]
#create a pie chart
plt.pie(sizes,labels=labels,autopct='%.2f%%',startangle=90,colors=['red','green','blue','yellow','cyan'])
#add title
plt.title('Pie chart for subject knowledge')
#display
plt.show()
plt.legend()
```

```python
#subplot() is used to create multiple subplots with in same figure

x=np.linspace(0,5,100)
y1=x**2
y2=np.sqrt(x)
y3=np.sin(x)
#create subplot
plt.figure(figsize=(20,20))#figsize(width,height)
#subplot1
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title('exponential graph')
plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title('Square root graph')

x2=np.linspace(0,5,100)
plt.subplot(2,2,3)
plt.plot(x2,y3)
plt.title('Sine graph')
y4=x2**3
plt.subplot(2,2,4)
plt.plot(x2,y4)
plt.title('cube graph')
plt.show()
```

```python
#invloves mplot3d plot_surface()

from mpl_toolkits.mplot3d import Axes3D
#create a 3d pplot
fig=plt.figure(figsize=(6,6)) # creates a new figure with size 6,6
ax=fig.add_subplot(111,projection='3d') #Adds a 3D subplot to the figure. The 111 indicates a single subplot in a 1x1 grid, and projection='3d' specifies a 3D plot.
#generation of sample data
x=np.linspace(-5,5,100) # Generates 1D arrays of 100 evenly spaced values from -5 to 5 for both x and y.
y=np.linspace(-5,5,100)
x,y=np.meshgrid(x,y) #Creates a 2D grid of x and y values using meshgrid.
z=np.sin(np.sqrt(x**2+y**2))
ax.plot_surface(x,y,z,cmap='viridis') #Plots the 3D surface using plot_surface. The cmap='viridis' argument specifies the color map
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

ax.set_title('sin function in 3 D')
plt.show()

```

```python
#create a 3d scatter plot
fig=plt.figure(figsize=(8,8))
ax=fig.add_subplot(111,projection='3d')
#generate data
x=np.random.rand(50)
y=np.random.rand(50)
a=np.random.rand(50)
b=np.random.rand(50)
c=np.random.rand(50)
z=np.random.rand(50)
#plot the scatter points
ax.scatter(x,y,z,color='r',marker='^')
ax.scatter(a,b,c,color='b',marker='o')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3d scatter plot')
plt.show()
plt.legend()
```
