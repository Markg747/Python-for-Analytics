```python
# conditional scatter plot
import numpy as np
import matplotlib.pyplot as plt

# Generate some sample data
np.random.seed(0) #random numbers generated using NumPy's random module will be the same each time you run your code,
x = np.random.randn(100)
y = np.random.randn(100)

# Scatter plot with condition
plt.scatter(x, y, c=np.where(x > 0, 'r', 'b'), label='x > 0')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Condition')
plt.legend()

# Show plot
plt.show()
```

```python
tips=sns.load_dataset('tips')
print(tips)
sns.boxplot(x='day',y='total_bill',data=tips,palette='pastel')
plt.xlabel('days of week')
plt.ylabel('Total bill')
plt.title("box plot")
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='swarm',palette='pastel')
plt.xlabel('days of week')
plt.ylabel('Total bill')
plt.title("categorical box plot")
```

```python
# doubt solution: importing data from sequel in python
import sqlite3

# Connect to the database
conn = sqlite3.connect('your_database.db')

# Create a cursor object
cursor = conn.cursor()

# Execute a query
cursor.execute('SELECT * FROM your_table_name')

# Fetch the results
results = cursor.fetchall()

# Print the results
for row in results:
    print(row)

# Close the connection
conn.close()
```

```python
# continution of above
import sqlite3
import pandas as pd

# Get the column names from the cursor description
columns = [description[0] for description in cursor.description]
#cursor.description attribute
# Column name
# Data type (as a constant from the database module, such as sqlite3.INTEGER or pymysql.FIELD_TYPE_VAR_STRING)
# Display size
# Internal size
# Precision
# Scale
# Nullability (True if the column can contain NULL values, False otherwise)

# Create a DataFrame
df = pd.DataFrame(results, columns=columns)

# Print the DataFrame
print(df)

# Close the connection
conn.close()
```

```python
# doubt solution :how to find  age through name without index no

import pandas as pd

# Sample DataFrame with names and ages
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
        'Age': [25, 30, 35, 40, 45]}
df = pd.DataFrame(data)

# Function to find age by name
def find_age_by_name(name):
    result = df[df['Name'] == name]['Age']
    if len(result) > 0:
        return result.values[0]
    else:
        return None

# Usage
name = 'Bob'
age = find_age_by_name(name)
if age is not None:
    print(f"The age of {name} is {age}.")
else:
    print(f"No record found for {name}.")
```
